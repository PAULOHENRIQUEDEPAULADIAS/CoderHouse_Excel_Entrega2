Através de uma tabela dinâmica, fica claro que **a maior parte das transações geradas provém de caixas eletrônicos, representando 34,35% do volume total.**

Através de uma tabela dinâmica, formatação condicional, e, para melhor visualização, geração de gráficos, pode-se concluir que **meses de Maio, Agosto, Setembro, Novembro e Dezembro, correspondem aos meses em que as pessoas mais recorrem a linhas de Créditos** e **no decorrer dos meses de Janeiro, Agosto, Setembro, Outubro e Dezembro, onde há o maior indício de gastos anuais em Débito**.<br>
Deduzimos que parte desses gastos são decorrentes de festividades, como Natal em Dezembro, ou contas anuais, como IPVA e IPTU em Dezembro e Janeiro, mas e os demais?

Através do uso de tabéla dinâmica e gráficos, identifico que, ***Estudantes* possuem o maior nível de gastos, tanto em Crédito quanto em Débito em relação as demais ocupações.** Sendo em ordem crescente, *Aposentados, Engenheiros, Doutores, e Estudantes*.
